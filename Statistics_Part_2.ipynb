1. What is hypothesis testing in statistics\*
--   
   Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves setting up two competing hypotheses (null and alternative) and determining which one is more supported by the sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis\*
 --
   The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default or starting assumption. The alternative hypothesis (H1 or Ha) is what you want to prove; it states that there is an effect or a difference.

3. What is the significance level in hypothesis testing, and why is it important\*
  --
   The significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true. It represents the threshold for determining statistical significance, commonly set at 0.05 (5%).

4. What does a P-value represent in hypothesis testing\*
   --
   The P-value is the probability of obtaining test results at least as extreme as the observed results, under the assumption that the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing\*
 --
* If the P-value ≤ α: reject the null hypothesis (evidence supports the alternative hypothesis).
* If the P-value > α: fail to reject the null hypothesis (not enough evidence against it).

6. What are Type 1 and Type 2 errors in hypothesis testing\*
--  
* Type I Error: Rejecting a true null hypothesis (false positive).
* Type II Error: Failing to reject a false null hypothesis (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing\*
--
* One-tailed test checks for an effect in one direction only (greater or less).
* Two-tailed test checks for an effect in both directions (different, without specifying direction).

8. What is the Z-test, and when is it used in hypothesis testing\*
--
   A Z-test is used when the population variance is known and the sample size is large (n > 30). It helps to determine if there is a significant difference between sample and population means.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing\*
--
   Z = (sample mean - population mean) / (population std. deviation / √(sample size))
   The Z-score tells how many standard deviations a data point is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution\*
--

   The T-distribution is used when the population standard deviation is unknown and/or the sample size is small (n < 30). It is wider and accounts for more variability in small samples.

11. What is the difference between a Z-test and a T-test\*
--
* Z-test is used when population standard deviation is known.
* T-test is used when population standard deviation is unknown and sample size is small.

12. What is the T-test, and how is it used in hypothesis testing\*
--
   The T-test is a statistical test used to compare the means of two groups (one-sample, two-sample, or paired). It determines whether the sample mean significantly differs from a known value or another sample.

13. What is the relationship between Z-test and T-test in hypothesis testing\*
 --
  Both are used to test hypotheses about population means. Z-test assumes known population standard deviation; T-test is used when this is unknown.

14. What is a confidence interval, and how is it used to interpret statistical results\*
--
   A confidence interval is a range of values derived from sample statistics that is likely to contain the true population parameter. It provides an estimate with a given level of confidence (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval\*
--
  Margin of error is the amount added/subtracted from the sample statistic to create the confidence interval. A larger margin increases the interval width, indicating more uncertainty.

16. How is Bayes' Theorem used in statistics, and what is its significance\*
--
  Bayes' Theorem updates the probability of a hypothesis based on new evidence. It is significant in decision-making, especially in medical testing, diagnostics, and machine learning.

17. What is the Chi-square distribution, and when is it used\*
--
  The Chi-square distribution is used in categorical data analysis, such as tests of independence and goodness-of-fit. It deals with variances and frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied\*
  --
  This test checks if a sample distribution matches an expected distribution. It compares observed and expected frequencies and calculates a Chi-square statistic.

19. What is the F-distribution, and when is it used in hypothesis testing\*
--
   
   The F-distribution arises in comparing two variances or in ANOVA tests. It is used when testing the ratio of two variances or comparing multiple group means.

20. What is an ANOVA test, and what are its assumptions\*
--
  ANOVA (Analysis of Variance) tests whether there are significant differences between the means of three or more groups. Assumptions include normality, independence, and equal variances.

21. What are the different types of ANOVA tests\*
--

* One-way ANOVA: tests effect of one factor.
* Two-way ANOVA: tests effect of two factors and their interaction.
* Repeated Measures ANOVA: for related groups or repeated measurements.

22. What is the F-test, and how does it relate to hypothesis testing?
   --
   The F-test evaluates whether two populations have different variances. In ANOVA, it tests the null hypothesis that all group means are equal using the F-distribution.

```python
#1 .Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
 #interpret the results
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05, two_tailed=True):
    """Perform Z-test comparing sample mean to population mean."""
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    if two_tailed:
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
        critical_value = norm.ppf(1 - alpha/2)
    else:
        p_value = 1 - norm.cdf(z_score)
        critical_value = norm.ppf(1 - alpha)

    print(f"Z-Score: {z_score:.4f}, P-Value: {p_value:.4f}")
    print(f"Critical Value: ±{critical_value:.4f}" if two_tailed else f"Critical Value: {critical_value:.4f}")
    print("Reject H₀" if p_value < alpha else "Fail to reject H₀")

# Example Usage
np.random.seed(42)
sample = np.random.normal(loc=5.1, scale=2.0, size=30)  # Simulated sample
z_test(sample, pop_mean=5.0, pop_std=2.0)
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    z = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
    p = 2 * (1 - norm.cdf(abs(z)))  # Correct calculation of two-tailed p-value
    return z, p

# Example usage
sample = [4.7, 5.1, 5.3, 4.9, 5.0, 4.8, 5.2]
z, p = one_sample_z_test(sample, pop_mean=5.0, pop_std=2.0)

print(f"Z-score: {z:.2f}")
print(f"P-value: {p:.4f}")
if p < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot@
import matplotlib.pyplot as plt

def plot_z_test(z_score, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)
    plt.plot(x, y, label='Standard Normal')
    plt.axvline(z_score, color='r', linestyle='--', label=f'Z-Score: {z_score:.2f}')

    # Critical regions
    cv = norm.ppf(1 - alpha/2)
    plt.axvline(cv, color='g', linestyle=':', label=f'Critical: ±{cv:.2f}')
    plt.axvline(-cv, color='g', linestyle=':')
    plt.fill_between(x, y, where=(x >= cv) | (x <= -cv), alpha=0.3, color='red')
    plt.legend()
    plt.show()

plot_z_test(z_score=2.1)  # Example
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing@
def plot_errors(true_mu=0, alpha=0.05, n=30, sigma=1):
    # H₀ distribution
    x = np.linspace(-4, 4, 1000)
    y_h0 = norm.pdf(x, loc=0, scale=sigma/np.sqrt(n))
    cv = norm.ppf(1 - alpha/2, scale=sigma/np.sqrt(n))

    # H₁ distribution (assuming true_mu ≠ 0)
    y_h1 = norm.pdf(x, loc=true_mu, scale=sigma/np.sqrt(n))

    plt.plot(x, y_h0, label='H₀ Distribution')
    plt.plot(x, y_h1, label='H₁ Distribution')

    # Type I Error (α)
    plt.fill_between(x, y_h0, where=(x >= cv) | (x <= -cv), alpha=0.3, color='red', label='Type I Error')

    # Type II Error (β)
    beta_cutoff = cv - true_mu
    plt.fill_between(x, y_h1, where=(x <= cv) & (x >= -cv), alpha=0.3, color='blue', label='Type II Error')

    plt.legend()
    plt.show()

plot_errors(true_mu=0.5)  # Example
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import scipy.stats as stats # Import the stats module

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(n)  # Standard error
    t_crit = stats.t.ppf((1 + confidence)/2, df=n-1)
    moe = t_crit * se  # Margin of Error
    return (mean - moe, mean + moe), moe

ci, moe = confidence_interval(sample)
print(f"95% CI: {ci}, Margin of Error: {moe:.4f}")
```

```python
#12 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data
def expected_frequencies(observed):
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    total = observed.sum()
    return np.outer(row_totals, col_totals) / total

print(expected_frequencies(observed))
```

```python
#19 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
#results

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data for three groups, each split into two levels of factor2
np.random.seed(42)
group1 = np.random.normal(loc=20, scale=3, size=30)  # Factor1 = A
group2 = np.random.normal(loc=22, scale=3, size=30)  # Factor1 = B
group3 = np.random.normal(loc=24, scale=3, size=30)  # Factor1 = C

# Create DataFrame
data = pd.DataFrame({
    'value': np.concatenate([group1, group2, group3]),
    'factor1': ['A']*30 + ['B']*30 + ['C']*30,
    'factor2': ['X']*15 + ['Y']*15 + ['X']*15 + ['Y']*15 + ['X']*15 + ['Y']*15
})

# Fit two-way ANOVA model with interaction
model = ols('value ~ C(factor1) + C(factor2) + C(factor1):C(factor2)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-way ANOVA results:\n", anova_table)

# Optional: Visualization using boxplot
sns.boxplot(x='factor1', y='value', hue='factor2', data=data)
plt.title("Interaction between Factor1 and Factor2")
plt.show()
```

```python
#27 Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import norm

def margin_of_error(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    z_score = norm.ppf(1 - (1 - confidence) / 2)
    se = std_dev / np.sqrt(n)
    moe = z_score * se
    return moe

# Example usage:
sample_data = [56, 60, 61, 63, 59, 62, 58, 57, 64, 60]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100:.0f}% confidence: ±{moe:.2f}")
```
