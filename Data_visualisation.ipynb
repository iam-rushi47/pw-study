```python
'''4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.'''
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D']
np.random.seed(0)

# Numerical values for each category
data = {
    'Category A': np.random.normal(50, 10, 100),  # Mean=50, Std=10
    'Category B': np.random.normal(60, 15, 100),  # Mean=60, Std=15
    'Category C': np.random.normal(45, 5, 100),   # Mean=45, Std=5
    'Category D': np.random.normal(55, 20, 100),  # Mean=55, Std=20
}

# Create a box plot to visualize the distribution
plt.boxplot([data['Category A'], data['Category B'], data['Category C'], data['Category D']],
            labels=categories,
            patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            whiskerprops=dict(color='blue'),
            flierprops=dict(markerfacecolor='red', marker='o', markersize=5))

# Add titles and labels
plt.title('Distribution of Numerical Values Across Categories')
plt.ylabel('Values')

# Show grid and plot
plt.grid(True)
plt.show()
```

```python
'''5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap'''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n = 100  # Number of samples
x = np.random.normal(0, 1, n)  # Feature 1
y = 0.8 * x + np.random.normal(0, 0.5, n)  # Feature 2 (correlated with x)
z = 0.5 * x + 0.5 * y + np.random.normal(0, 0.7, n)  # Feature 3 (correlated with x and y)

# Create a DataFrame
data = pd.DataFrame({'Feature 1': x, 'Feature 2': y, 'Feature 3': z})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

# Bokeh Assignment Solution


```python
''' 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Use notebook output (use output_file("sine_wave.html") for file output)
output_notebook()

# Generate x and y values
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Plot the sine wave
p.line(x, y, line_width=2, color='navy', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
'''2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.'''
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Show plot inside notebook (or use output_file(...) for HTML)
output_notebook()

# Generate random data
np.random.seed(42)
data = {
    'x': np.random.rand(50) * 100,
    'y': np.random.rand(50) * 100,
    'sizes': np.random.randint(10, 30, 50),
    'colors': np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], 50)
}

df = pd.DataFrame(data)
source = ColumnDataSource(df)

# Create Bokeh figure
p = figure(title="Bokeh Scatter Plot with Variable Size and Color",
           x_axis_label='X', y_axis_label='Y', width=700, height=400)

# Scatter plot with sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)

```

```python
''' 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset. '''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Render inside notebook
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label="Fruit", y_axis_label="Count",
           width=600, height=400)

# Bar chart
p.vbar(x='fruits', top='counts', width=0.5, source=source,
       color=factor_cmap('fruits', palette='Spectral4', factors=fruits))

# Styling
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show plot
show(p)
```

```python
'''4. Create a Bokeh histogram to visualize the distribution of the given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)'''
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Show output in notebook
output_notebook()

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           width=700, height=400)

# Draw histogram using quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show plot
show(p)

```

```python
'''5. Create a Bokeh heatmap using the provided dataset.data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


'''

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256

# Display in notebook
output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten data for ColumnDataSource
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Define color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Heatmap",
           x_axis_label="X", y_axis_label="Y",
           tools="hover", tooltips=[("Value", "@value")],
           match_aspect=True,
           width=600, height=500)

# Draw heatmap using rect glyph
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=transform('value', color_mapper), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), label_standoff=12)
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
