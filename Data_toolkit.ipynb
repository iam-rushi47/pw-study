3. Explain the following terms

   1.The difference in np.array, np.asarray and np.asanyarrayX

   2 The difference between Deep copy and shallow copy

 --
1. Differences between np.array, np.asarray, and np.asanyarray
Function	Behavior	Use Case
 -  np.array():-	1)Always creates a new copy of the input data (even if it's already a NumPy array).

     2)	Use when you need a guaranteed new array (e.g., modifying without affecting the original).

 -  np.asarray():-	1)Converts input to an array only if it isn't already one.

       2)If the input is already a NumPy array, it does not create a copy (unless dtype is changed).

     3)	Use when you want to ensure an array but avoid unnecessary copies.

 - np.asanyarray():-1)Similar to asarray, but allows subclasses (e.g., matrix, masked arrays) to pass through unchanged.

       2)	Use when you need to preserve subclass types (like np.matrix).
Example:

    
    import numpy as np

    a = np.array([1, 2, 3])

    np.array() → Always new copy
    b = np.array(a)  # New memory allocated
    b[0] = 99  # Does not affect 'a'

    np.asarray() → No copy if already an array
    c = np.asarray(a)  # Shares memory with 'a'
    c[0] = 100  # Modifies 'a' as well

    np.asanyarray() → Allows subclasses
    m = np.matrix([1, 2, 3])
    d = np.asanyarray(m)  # Still a matrix (not converted to ndarray)

2. Difference between Deep Copy and Shallow Copy
 - Shallow Copy:-
   
   1)Shallow Copy	Creates a new object, but shares references to the same underlying data.

   2) Modifying one affects the other.

   3)Use when you need a view of the original data (memory efficient).

 - Deep Copy:-

  1)Deep Copy	Creates a completely independent copy with its own memory.

  2) Changes do not affect the original.
  
  3)	Use when you need a fully independent duplicate.

**Example (NumPy Arrays)**:

    import numpy as np

    Original array
    arr = np.array([1, 2, 3])

    # Shallow Copy (view)
    shallow_copy = arr.view()
    shallow_copy[0] = 99  # Modifies 'arr' as well

    # Deep Copy (independent)
    deep_copy = arr.copy()
    deep_copy[0] = 100  # Does not affect 'arr'

```python
'''4. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array'''
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

print("Original Array:")
print(random_array)

# a) Extract all even integers
even_numbers = random_array[random_array % 2 == 0]

# b) Extract all odd integers
odd_numbers = random_array[random_array % 2 != 0]

print("\nEven Numbers:")
print(even_numbers)

print("\nOdd Numbers:")
print(odd_numbers)
```

```python
'''15. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:


a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series DataFrame
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.uniform(15, 35, size=30)  # Random temperatures between 15°C and 35°C
humidity = np.random.uniform(40, 90, size=30)     # Random humidity between 40% and 90%

df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set 'Date' as the index for plotting
df.set_index('Date', inplace=True)

# a) Plot Temperature and Humidity with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on the left y-axis
ax1.plot(df.index, df['Temperature'], color='red', label='Temperature (°C)')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], color='blue', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Label x-axis
ax1.set_xlabel('Date')

# c) Set title
plt.title('Temperature and Humidity Over Time')

# Optional: Grid and layout
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
''' 16.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib
:a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution (mean=0, std=1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Calculate the PDF and overlay it
pdf = norm.pdf(bins, loc=0, scale=1)
plt.plot(bins, pdf, 'r--', linewidth=2, label='Normal PDF')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Show the plot
plt.show()
```

```python
''' 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.'''
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays (x and y)
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def get_quadrant(x_val, y_val):
    if x_val > 0 and y_val > 0:
        return 'Quadrant I'
    elif x_val < 0 and y_val > 0:
        return 'Quadrant II'
    elif x_val < 0 and y_val < 0:
        return 'Quadrant III'
    elif x_val > 0 and y_val < 0:
        return 'Quadrant IV'
    else:
        return 'On Axis'

# Create a DataFrame
df = pd.DataFrame({
    'x': x,
    'y': y,
    'Quadrant': [get_quadrant(x[i], y[i]) for i in range(len(x))]
})

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', s=80, edgecolor='black')

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
''' 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'8  .  '''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Ensure output is shown in the notebook
output_notebook()

# Generate sine wave data
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create the figure
p = figure(title="Sine Wave Function", x_axis_label='X Axis', y_axis_label='Y Axis', width=800, height=400)

# Plot the sine wave
p.line(x, y, line_width=2, color='navy', legend_label='sin(x)')

# Show plot in notebook
show(p)

```

```python
'''19.  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'8
'''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Show Bokeh output in notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))

# Assign colors based on value (e.g., gradient from light to dark blue)
from bokeh.transform import linear_cmap
from bokeh.palettes import Blues5

# Create data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value', width=700, height=400, tools="")

# Add bars with color mapped to values
mapper = linear_cmap(field_name='values', palette=Blues5, low=min(values), high=max(values))
p.vbar(x='categories', top='values', width=0.6, source=source, color=mapper, line_color='black')

# Add hover tool
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Show the plot
show(p)

```

```python
''' 20.  Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'8'''

import plotly.graph_objs as go
import numpy as np

# Generate random data
x = np.arange(0, 20)
y = np.random.randn(20)

# Create line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Random Line'))

# Add labels and title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    template='plotly_white'
)

# Show plot
fig.show()
```
